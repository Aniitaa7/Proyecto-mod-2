```python
#sacamos los resultados canciones(track) para Género rock. Tiene un límite de 50 resultados
results = sp.search(q=f'genre:rock', type='track', limit=50)

results #el resultado ya es un diccionario
```

```python
# Resultados en el rango de fechas 2000-2024

results_in_range = [] 
        
for item in results['tracks']['items']:
    relase_date = item ['album']['release_date'] # Creamos la variable para fecha lanzamiento
    if '2000-01-01'<= relase_date <= '2024-12-31': # Le damos el rango de fecha 
        results_in_range.append(item)

# Hacemos un diccionario con todos los valores in range  
  
artistas = []
canciones = []
año_lanzamiento = []    
        
canciones_rock = {'Artista':artistas, 'Genero_musical': 'Rock', 'Tipo': 'Cancion', 'Titulo_cancion':canciones, 'Año_lanzamiento':año_lanzamiento}

for item in results_in_range:

    #para extraer artistas
    artistas. append (item ['album']['artists'][0]['name'])
   

    #para extraer canciones 
    canciones. append (item ['name'])

    # Para extraer año lanzamiento
    año_lanzamiento. append (item ['album']['release_date'][0:4]) #Sólo el año
                 
# Sacamos los datos limpios en un DataFrame
df_canciones_rock = pd.DataFrame(canciones_rock)
df_canciones_rock # canciones_rock es un diccionario. Por tanto es un json?????

```
