```python
# Primera inserción: tabla de artistas_spotify

cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                              host='127.0.0.1',
                              database='music_sp_mb_lfm')

# Crear cursor
mycursor = cnx.cursor()

# SQL para insertar un artista
sql_insertar_artista = "INSERT INTO artistas_spotify (artista) VALUES (%s)"

# Iterar sobre la lista de artistas y verificar si ya existe
for artista in tuplas_lista_artistas_id:
    # Verificar si el artista ya existe en la tabla
    mycursor.execute("SELECT COUNT(*) FROM artistas_spotify WHERE artista = %s", (artista[0],))
    count = mycursor.fetchone()[0]
    
    if count == 0:
        # Si no existe, insertar el artista
        mycursor.execute(sql_insertar_artista, (artista[0],))
        print(f"{artista[0]} insertado.")
    else:
        print(f"{artista[0]} ya existe. No se inserta.")

# Confirmar los cambios en la base de datos
try:
    cnx.commit()
    print("Cambios confirmados en la base de datos.")
except mysql.connector.Error as err:
    print("Error:", err)
    print("Error Code:", err.errno)
    print("SQLSTATE", err.sqlstate)
    print("Message", err.msg)

# Cerrar el cursor y la conexión
mycursor.close()
cnx.close()

```

# INSERCIÓN DE LOS DATOS DE MUSICBRAINZ

```python
# Convertir el csv a dataframe
df_musicbrainz_id = pd.read_csv('SQL_musicbrainz_id.csv')
```

```python
#Cambio de nombre a las columnas para que no haya error
df_musicbrainz_id.rename(columns={'Nombre': 'id_artista', 'Tipo': 'tipo', 'País de Origen': 'pais_origen', 
                                               'Área de Origen': 'area_origen', 'Inicio de actividad': 'inicio_actividad',
                                                'Fin de actividad': 'fin_actividad', 'Fecha nacimiento': 'fecha_nacimiento',
                                                'Fecha fallecimiento': 'fecha_fallecimiento'}, inplace=True)

print(df_musicbrainz_id)
```

```python
# Reecuento de los 'NaN' en todas las columnas
nan_counts = df_musicbrainz_id.isna().sum()
print(nan_counts)
```

```python
# Reemplazar NaN por None en todo el DataFrame
df_musicbrainz_id=df_musicbrainz_id.where(pd.notna(df_musicbrainz_id), None)
```
