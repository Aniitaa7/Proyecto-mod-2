```python
import requests
import pandas as pd
# Configuración para poder visualizar todas las columnas de los DataFrames
pd.set_option('display.max_columns', None) 
import json
#para poder retrasar las llamadas
import time
```

```python
# leemos el archivo csv con pandas
df_lista_artistas = pd.read_csv("lista_artistas.csv")
df_lista_artistas
```

```python
# hacemos una lista con el dataframe que nos devuelve
artistas = list(df_lista_artistas['0'])
artistas
```

```python
info_artistas = {'Nombre':[],'Tipo':[], 'País de Origen':[], 'Área de Origen':[], 'Inicio de actividad':[], 'Fin de actividad':[], 'Fecha nacimiento':[], 'Fecha fallecimiento':[]}

for artista in artistas:
    time.sleep(1)
    url = f'https://musicbrainz.org/ws/2/artist/?query={artista}&fmt=json'
    datos_artista = requests.get(url)
    print(datos_artista.status_code) #los errores son 503 Hay que ver lo de retrasar las llamadas
    data = datos_artista.json()
    # Daba un error si no encontraba el artista. Así que antes comprueba que exista el dato artista, que no esté vacío
    if ('artists') in data and len(data['artists']) > 0:
        
            try:
                nombre_artista = data['artists'][0]['name']
                info_artistas['Nombre'].append(nombre_artista)
            except (KeyError, ValueError):    
                info_artistas['Nombre'].append('Error')
            
            try:
                tipo_artista = data['artists'][0]['type']
                info_artistas['Tipo'].append(tipo_artista)
            except (KeyError, ValueError):     
                info_artistas['Tipo'].append('Error') 
            
            try:
                pais_artista = data['artists'][0]['area']['name']
                info_artistas['País de Origen'].append(pais_artista)
            except (KeyError, ValueError):      
                info_artistas['País de Origen'].append('Undefined')         
            
            try:
                area_artista = data['artists'][0]['begin-area']['name']
                info_artistas['Área de Origen'].append(area_artista)
            except (KeyError, ValueError):    
                info_artistas['Área de Origen'].append('Undefined')        

                
            # Si es un grupo pondrá los datos en inicio y fin de actividad y los otros undefined

            # También daba error al encontrar type vacío
            data_type = data['artists'][0]    
            if data_type.get('type') == 'Group': # con .get no da error, pone None
                try:
                    inicio_actividad = data['artists'][0]['life-span']['begin']
                        
                    info_artistas['Inicio de actividad'].append(inicio_actividad)
                    info_artistas['Fecha nacimiento'].append('-')

                    fin_actividad = data['artists'][0]['life-span']['ended'] 

                    info_artistas['Fin de actividad'].append(fin_actividad)
                    info_artistas['Fecha fallecimiento'].append('-')
                        
                    
                except (KeyError, ValueError):  
                    info_artistas['Inicio de actividad'].append('Undefined')
                    info_artistas['Fecha nacimiento'].append('Undefined')
                    info_artistas['Fin de actividad'].append('Undefined')
                    info_artistas['Fecha fallecimiento'].append('Undefined')                   

                # Si es una persona pondrá los datos en nacimiento y fallecimiento y los otros undefined
            else:
                try:
                    inicio_actividad = data['artists'][0]['life-span']['begin']
                        
                    info_artistas['Inicio de actividad'].append('-')
                    info_artistas['Fecha nacimiento'].append(inicio_actividad)

                    fin_actividad = data['artists'][0]['life-span']['ended'] 

                    info_artistas['Fin de actividad'].append(fin_actividad)
                    info_artistas['Fecha fallecimiento'].append('-')
                        
                    
                except (KeyError, ValueError):  
                    info_artistas['Inicio de actividad'].append('Undefined')
                    info_artistas['Fecha nacimiento'].append('Undefined')
                    info_artistas['Fin de actividad'].append('Undefined')        
                    info_artistas['Fecha fallecimiento'].append('Undefined')
               
        
else: # Si artist está vacío pasa al siguiente artista
    pass
info_artistas
```

```python
# Verificar longitudes (ChatGpt)
for key, value in info_artistas.items():
    print(f"{key}: {len(value)}")
```

```python
# Convertimos a un DataFrame
df_info_artistas = pd.DataFrame(info_artistas)
```
