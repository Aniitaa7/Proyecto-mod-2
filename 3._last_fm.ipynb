```python
if datos_artista.status_code == 200:
        data = datos_artista.json()
        artist_info = data.get('artist', {})
        
        if artist_info:
            info_artistas['Nombre'].append(artist_info.get('name', 'N/A'))
            info_artistas['Biografia'].append(artist_info.get('bio', {}).get('summary', 'N/A'))
            info_artistas['Listeners'].append(artist_info.get('listeners', 'N/A'))
            info_artistas['Playcount'].append(artist_info.get('playcount', 'N/A'))
            info_artistas['Artistas similares'].append([similar['name'] for similar in artist_info.get('similar', {}).get('artist', [])])
        else:
            print(f"No se encontró información para el artista: {artista}")
else:
        print(f'Error al obtener información para {artista}: {datos_artista.status_code}')
```

```python
info_artistas_df = pd.DataFrame(info_artistas)
print(info_artistas_df)
```
